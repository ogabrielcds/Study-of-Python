# Listas

```python
###Exercício 1

# input:
vec = input().split()
vec = [int(x) for x in vec]

# soma dos vetores:
s = 0

for x in range(len(vec)):
    s+=vec[x]
print(s)

# número de ocorrências:
l = []

for x in vec:
    if x == vec[-1]:
        l.append(x)
print(len(l))

# máximo da lista:
print(max(vec))

#indices dos máximos:
i = list()
for indice, item in enumerate(vec):
    if item == max(vec):
        i.append(indice)
print(*i)
```

```python
### Exercício 2

# input:
vec = []

while True:
    i = input()
    i = int(i)
    if i < 0:
        break
    else:
        vec.append(i)

# qnt de valores > 5:
qnt_cinco = []
for x in vec:
    if x > 5:
        qnt_cinco.append(x)
c = len(qnt_cinco)
print(c)
# soma dos valores pares e valores ímpares:
s_p = 0
s_i = 0
for x in vec:
    if x%2 == 0:
        s_p += x
    else:
        s_i += x
print(s_p)
print(s_i)

# qnt total de valores:
comp = len(vec)
print(comp)

# o vetor:
print(*vec)
```

```python
### Exercício 3

n = [float(x) for x in input().split()]

s = n[0]
p = n[0]

for i in range(1,len(n)):
    print('%.2f' % s, '%.2f' % p)
    s += n[i]
    p *= n[i]
print('%.2f' % s, '%.2f' % p)
```

# Tuplas

### 1) Exercício para escrever por extenso números de 0 a 20, utilizando tuplas.

```python
#Meu código:

num = int(input('Digite um número entre 0 e 20:'))

resp = ''
tupla = ('zero', 'um', 'dois', 'três', 'quatro', 'cinco', 'seis', 'sete', 'oito', 'nove', 'dez', 'onze', 'doze', 'treze', 'catorze', 'quinze', 'dezesseis', 'dezessete', 'dezoito', 'dezenove', 'vinte')

if 20 >= num >= 0:
 resp = tupla[num]
 print('Você digitou o número {}.'.format(resp)) 
else:
 print('Tente novamente! Digite um número de 0 a 20')
```
