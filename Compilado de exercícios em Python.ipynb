### 1) Exercício para escrever por extenso números de 0 a 20, utilizando tuplas.

```python
#Meu código:

num = int(input('Digite um número entre 0 e 20:'))

resp = ''
tupla = ('zero', 'um', 'dois', 'três', 'quatro', 'cinco', 'seis', 'sete', 'oito', 'nove', 'dez', 'onze', 'doze', 'treze', 'catorze', 'quinze', 'dezesseis', 'dezessete', 'dezoito', 'dezenove', 'vinte')

if 20 >= num >= 0:
 resp = tupla[num]
 print('Você digitou o número {}.'.format(resp)) 
else:
 print('Tente novamente! Digite um número de 0 a 20')
```

### 2) Exercício sobre dicionários

```python
# Ok - Resolução Ex 1:

UNIDADES = { 0: '', 1: 'I', 2: 'II', 3: 'III', 4: 'IV', 5: 'V', 6: 'VI', 7: 'VII', 8: 'VIII', 9: 'IX' }
DEZENAS = { 0: '', 1: 'X', 2: 'XX', 3: 'XXX', 4: 'XL', 5: 'L', 6: 'LX', 7: 'LXX', 8: 'LXXX', 9: 'XC' }
CENTENAS = { 0: '', 1: 'C', 2: 'CC', 3: 'CCC', 4: 'CD', 5: 'D', 6: 'DC', 7: 'DCC', 8:'DCCC', 9:'CM' }

num = input()
if len(num) == 3: print(CENTENAS[int(num[0])] + DEZENAS[int(num[1])] + UNIDADES[int(num[2])])
elif len(num) == 2: print(DEZENAS[int(num[0])] + UNIDADES[int(num[1])])
else: print(UNIDADES[int(num[0])])
    
# Resolução Ex 2:

frase = input().lower()
alfabeto = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
dic = {'vogais': 0,'consoantes': 0}
vogais = {} 
consoantes = {} 
rep = [] #achei mais fácil fazer isso do que fazer do jeito certo, mim desculpa

for i in frase: 
    if i in alfabeto:
        if i == 'a' or i=='e' or i=='i' or i=='o' or i=='u':
            if i not in rep:
                rep.append(i)
                vogais[i]=1 
            else:
                vogais[i]+=1
        else:
            if i not in rep:
                rep.append(i)
                consoantes[i]=1
            else:
                consoantes[i]+=1 
    else:
        pass

dic['vogais'] = vogais
dic['consoantes'] = consoantes 
print(dic)

# Resolução Ex 3:

nota_alumni = {}
numA = int(input())

for i in range (numA):
    aluno_bruto = input().split() 
    somatorio = 0
    for j in range (1, 11):
        somatorio += float(aluno_bruto[j]) 
    nota = (somatorio/10) * float(aluno_bruto[-1]) 
    nota_alumni[aluno_bruto[0]] = nota 

numDisplay = int(input()) 
disp = [] 
for i in range (numDisplay):
    nome = input() 
    disp.append(nome)

for i in disp:
    print(i + ': ' + str("%.2f"%(nota_alumni[i])))
```
