# Tuplas


    As tuplas servem para trabalhar com mais de uma variável, isto é, estruturas compotas, lembrando que definimos uma variável como um lugar na memória que recebe uma atribuição e, ao atribuir um valor, o lugar na memória é substituído pela nova atribuição. 

    As tuplas, no entanto, são imutáveis.

    As tuplas, guardam mais de um valor e, além disso, podem ser de diferente classes, como strings, int, float, dentre outros.

    Por fim, lembrar que existem  as seguintes estruturas compostas: tuplas (), listas [] e dicionários {} e estas são as respectivas sintaxes.

```python
for pos, membro in enumerate(tupla):
    print(f'O seguinte membro pertence a uma classe: {membro} e está na posição {pos}')
```

# Dicionários

    Os dicionários implementam uma estrutura conhecida como mapeamento, isto é, uma coleção de associações entre chave (Key) e valor (value).
    
    A chave é semelhante ao índece, ou seja, basta dar colchete chave para localizar um valor dentro do dicionário, sendo que a chave pode ser string, int e etc.
    
    O conteúdo ou valor, por sua vez, pode ser qualquer coisa, inclusive listas ou outro dicionário.
    
    Os dicionários são definidos por chaves da seguinte maneira:
    D = {chave_1: valor_1,..., chave_n: valor_n}
    
    Dicionários vazios:
    D = {}



    Exemplo da criação de agenda:

```python
#Nesse exemplo a chave seria os nomes e os valores seriam os telefones:

agenda = {'Paula': [985217463], 'Ana': [999997581,988567321], 'Pedro': [925256410]}
tel_ana = agenda['Ana']
print(tel_ana)
```

### Método copy
    Retorna uma cópia do dicio com os mesmos pares chave-valor. Sintaxe: dicio2 = dicio1.copy()

```python
# Análise do método copy na passagem por referência da memória:

dic1 = {'João': 10, 'Maria': 9}
dic2 = dic1.copy()
dic2['Ana'] = 15
dic1['João'] = 9

print(dic1)
print(dic2)
```

```python
# Análise do método copy se o conteúdo for lista ou dicionário:
#Repare que, nesse caso, temos novamento o problema da passagem por referência:

dic1 = {'X': [1,2], 'Y': [3,4]}
dic2 = dic1.copy()

dic2['Z'] = [5,6]
dic1['X'] += [3]

print(dic1)
print(dic2)

#Para resolver esse problema, devemos usaro módulo copy a seguir
```

### Módulo copy
    Uma alternativa ao método copy é a função deepcopy. Sintaxe: dicio2 = copy.deepcoopy(dicio1).

```python
#Repare que nesse caso, os dicios ficam independemtes, podendo fazer as alterações isoladamente.

import copy

dic1 = {'X': [1,2], 'Y': [3,4]}
dic2 = copy.deepcopy(dic1)
dic1['X'] += [13]

print(dic1)
print(dic2)
```

### Método fromkeys
    Esse método retorna um novo dicionário cujo valores são iguais para cada um das chaves especificadas. Se os valores não forem especificados, então atribuirá o valor 'None' para cada chave. Sintaxe: dicio.fromkeys([chave_1, chave_2], valor)

```python
A = {}.fromkeys([2,3])
print(A)

D = dic1.fromkeys(['X', 'Y'], 'Gabriel')
print(D)
```

### Método get
    Retorna o valor associado à chave (Ex.1). Posso dar '.get' em um valor que não existe dentro do dicio e atribuir um valor específico depois da vírgula (Ex.3). Se não atribuir nada, retorna 'None'(Ex.2), da seguinte maneira:

### Método items
    Esse método é o 'pai' dos métodos keys e values, pois retorna uma lista com todos os pares chave-valor do dicionário, da seguinte maneira:
    
    Além disso, lembre-se que para obter a lista é necessário dar o comando 'list(dicio.items())', pois 'dicio.itens()' retornará apenas dict_items(list), como mostrado abaixo:

```python
D = {'X': [1,2], 'Y': [3,4]}

D.items()

print(D.items())
print(list(D.items()))
```

### Método Keys e método Values:
    Retorna uma lista com todas as chaves do dicionário. Sintaxe: dicio.keys()
    No entanto, para transformar em uma lista de fato, é necessário utilizar um comando chamado **list(dicio.keys())**,pois o python retornará o dict_keys(), segue o exemplo:
    
    Analogamente, podemos utilizar o método 'values' para retornar uma lista de todos valores do dicionário. Novamente, a saída será dict_values(). Dessa forma, é preciso ultilizar o comando list(dicio.values()), da seguinte maneira:

```python
d = {'X': [1,2], 'Y': [3,4]}
print(d.keys())

print(list(d.keys()))

print(list(d.values()))
```

### Médoto Pop
    Retorna o valor correspondente á chave e, ao mesmo tempo, remove o par chave-chave do dicio. Sintaxe: dicio.pop(chave)

```python
#Repare que excluio a chave 'Y' e o seu respectivo valor, [3,4].

d = {'X': [1,2], 'Y': [3,4]}
d.pop('Y')
print(d)

# Além disso, é possível renomear uma chave de um dicio, utilizando esse conceito:

d = {'um': 1, 'DOIS': 2}

d['UM'] = d.pop('um')
print(d)
```

### Método popitem
    Esse método remove um par chave-valor ALEATÓRIO do dicionário. Sintaxe: dicio.popitem()

```python
d = {'X': [1,2], 'Y': [3,4]}

print(d.popitem()) #Nesse caso printará o valor que está sendo excluído

print(d) #O dicionário atualizado. 
```

### Comando del
    Esse comando exclui um par chave- valor específico do dicionário. Sintaxe: del dicio[chave].

```python
D = dict(X=1, Y=2, Z=3)
del D['Y']

chaves = D.keys()
valores = D.values()

print(D) #Printou o dicio sem o par chave-valor ('Y': 2), pois foi usado o del.
print(list(chaves)) #Printou a lista atualizada de chaves
print(valores) #Printou a lista atualizada de valores, porém no formato sem o list.
```

### Método Update
    Atualiza um dicionário com os elementos de outro dicionário, isto é, concatena os dicionários. Sintaxe: dicio1.update(dicio2)

```python
dic1 = {'X': 1, 'Y': 2, 'Z': 3}
dic2 = {'A': 4, 'B': 5, 'C': 6}

dic1.update(dic2)

print(dic1) #Repare que o dic1 recebeu os pares chave-valor do dic2.
```

    Outra forma de utilizar esse comando é especificando os pares chaves-valores que devem ser atualizados, da seguinte maneira

```python
dic0 = {'X': 1, 'Y': 2, 'Z': 3, 'A': 4, 'B': 5, 'C': 6}
dic0.update(Y=7, W=11) # Foi especificado o par chave-valor dentro dos parênteses.

print(dic0)
```

    Outra forma de usar esse comando, é atualizando um dicionário com outro dicionário da seguinte forma:

```python
dic1 = {'X': 1, 'Y': 2, 'Z':3} #Defini o dicio1
dic2 = {'W': ['a']} # Defini o dicio2
dic1.update(dic2) # Atualizei o dicio1, que agora contém o dois
dic2 ['W'] += ['b'] # Acrescentei estes valores ao dicio2
print(dic1) # Repare que aqui também acontete passagem por referência, pois é lista
```

```python
# O mesmo exemplo sem passagem de referência, isto é, sem listas:
dic1 = {'X': 1, 'Y': 2, 'Z':3} #Defini o dicio1
dic2 = {'W': 'a'} # Defini o dicio2
dic1.update(dic2) # Atualizei o dicio1, que agora contém o dois
dic2 ['W'] += 'b' # Acrescentei estes valores ao dicio2
print(dic1)
print(dic2)
#Os resultados são diferentes
```

### Iterando com dicionários:
    Em dicionários mão temos uma ordem definida. Além disso, a iteração occore a partir das chaves, isto é, os mecanismos de repetição vão percorrer as chaves, da seguinte maneira:

```python
notas = {'João': [3,2], 'Ana': [7.5,9], 'Gabriel': [2.17,10,10]}

for nome in notas:
    media = 0.75*(sum(notas[nome])/len(notas[nome]))
    print('A média das provas de', nome, 'é', media)
```

```python
notas = {'João': [3,2], 'Ana': [7.5,9], 'Gabriel': [2.17,10,10]}

for nome in notas:
    media = 0.75*(sum(notas[nome])/len(notas[nome]))
    print(f'A média das provas de {nome} é {media:.2f}')
#Repare na formação da média, para strings formatadas o arredondamento se dá dessa forma.
```

### Iterando com o médoto items()
    Lembre-se que esse método items é o método pai do keys() e values() e retorna o par chave-valor do dicio.

```python
idades = {'Pedro': 23, 'Chico': 29, 'Ana': 30, 'Lúcia': 68, 'Juca': 56}

for nome, idade in idades.items():
    print(f'{nome} tem {idade} anos')
```

### Iterando com o método keys() e items()

    Analogamente, é possível utilizar os métodos keys() e values()

```python
idades = {'Pedro': 23, 'Chico': 29, 'Ana': 30, 'Lúcia': 68, 'Juca': 56}

for nome in idades.keys():
    print(nome)

for idade in idades.values():
    print(idade)
    
# Nesse exemplo, separa as chaves dos valores
```

### Iterando com 'sorted + keys' e 'sorted + values'
    Esse método organiza as saídas em ordem alfabética. Pode ser combinado com o método keys, values e items.

```python
idades = {'Pedro': 23, 'Chico': 29, 'Ana': 30, 'Lúcia': 68, 'Juca': 56}

for nome in sorted(idades.keys()):
    print(nome)
    
for idade in sorted(idades.values()):
    print(idade)    
```

### Iterando sobre valores únicos 'set + values'
    O comando set irá remover os valores repetidos do dicionário da seguinte maneira:

```python
idades = {'Pedro': 23, 'Chico': 29, 'Ana': 30, 'Lúcia': 68, 'Juca': 56, 'Julia': 23}

for idade in sorted(set(idades.values())): 
    print(idade)

# Além de estar em ordem, o programa está excluindo os caracteres repetidos.
```

```python
usuarios = {'aeinstein': {'nome': 'albert', 
                         'sobrenome': 'einstein', 
                         'universidade': 'princeton'}, 
           'kthorne': {'nome': 'kip', 
                       'sobrenome': 'thorne', 
                       'universidade': 'caltech'}}
# Na iteração, para acessar um dicionário dentro de um dicionário, usa-se o comando 'info'.

for username, info in usuarios.items():
    print('Username:', username)
    nome_completo = info['nome']+' '+ info['sobrenome']
    local = info['universidade']
    print('\tNome completo:', nome_completo.title())
    print('\tUniversidade:', local.title())
```

# Funções

    A sintaxe para construir as funções é constituída de um bloco de definição e, após isso, um bloco responsável por chamar essa função 

```python
#Códigos para calcular a média:

#Bloco de definição:
def media (a,b):
    m = (a + b)/2
    return m

#Bloco responsável por chamar a função:
m = media(50,100)
print(f'A média é {m}')
```

```python
#Código da suma de números:

#Definição da função:
def minha_soma (x_min, x_max, inc):
    soma = 0
    for x in range (x_min, x_max, inc):
        soma += x
    return soma

#Primeira forma de retornar a função
S = minha_soma(0,5,1)
print(S)

#Segundo forma de retornar a função com exemplo de incremento diferente de 1:
print(minha_soma(3,9,2))
```

    Além disso, as funções podem returnar múltiplos valores. Para isso, pode-se criar múltiplas variáveis, afim dereceber os valores. 

```python
#Código para calcular a média e o desvio de um vetor

V = [1.1,9.2,0.7,3.9,8.4]

def stat2(vetor):
    n, desvio = len(vetor),0
    media = sum(vetor)/n
    
    for i in range(n):
        desvio += (vetor[i] - media)**2
    desvio = (desvio/n)**0.5
    
    return media, desvio

m,d = stat2(V)
print('media = %.4f desvio = %.4f' % (m,d))
```
